# Linear Regression Practice

Implementasi algoritma Linear Regression menggunakan Scikit-Learn.

**Studi Kasus:**
Hubungan Jam Belajar vs Nilai Siswa (Student Scores).

**Tech Stack:**
- Python
- Pandas (Data processing)
- Matplotlib (Visualization)
- Scikit-Learn (Model Fitting)

**Hasil:**
Mencapai akurasi (R^2 Score) sebesar 95%

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from google.colab import drive
```

```python
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/DataSet/student_scores.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
plt.scatter(df['Hours'], df['Scores'])
plt.xlabel('Hours')
plt.ylabel('Scores')
```

```python
reg = LinearRegression()
reg.fit(df[['Hours']], df['Scores'])
```

```python
reg.predict(pd.DataFrame({'Hours' : [9.5]}))
```

```python
akurasi = reg.score(df[['Hours']], df['Scores'])
print(f"Akurasi Model (R^2): {akurasi}")
```
